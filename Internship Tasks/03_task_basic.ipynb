Implement KNN (K-Nearest Neighbours) Classifier

```python
#importing the dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
#loading the dataset into pandas dataframe
from sklearn.datasets import load_digits
digits = load_digits()

digits_df = pd.DataFrame(digits.data)
```

```python
#first 5 rows of dataset
digits_df.head()
```

```python
#rows and column of the dataset
digits_df.shape
```

```python
#creating the target column
digits_df["target"] = digits.target
```

```python
#labelling the data 
x = digits_df.drop(columns="target",axis=1)
y = digits_df["target"] = digits.target
```

```python
#splittiing the data in training and testing data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3 ,random_state=3,stratify=y)
print(x.shape,x_train.shape,x_test.shape)
```

```python
#training the model 
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

```python
#prediction using training data
x_train_predict = model.predict(x_train)
x_train_accuracy = accuracy_score(x_train_predict,y_train)
print("Accuracy of the train data: ", round(x_train_accuracy*100,2),"%")
```

When n_neighbours = 3 , The train data accuarcy = 99.4%

When n_neighbours = 5 , The train data accuarcy = 98.97%

When n_neighbours = 10 , The train data accuarcy = 98.33%

```python
#prediction using testing data
x_test_predict = model.predict(x_test)
x_test_accuracy = accuracy_score(x_test_predict,y_test)
print("Accuracy of the test data: ",round(x_test_accuracy*100,2) ,"%")
```

When n_neighbours = 5 , The test data accuarcy = 98.7%

When n_neighbours = 3 , The test data accuarcy = 98.33%

When n_neighbours = 10 , The test data accuarcy = 98.33%


```python
#evaluating the model using precision , recall and f1_score
def precision_recall_f1score(true,pred):

    precision_value = precision_score(true,pred , pos_label=1 , zero_division=0,average="weighted")
    recall_value = recall_score(true,pred,pos_label=1,zero_division=0,average="weighted")
    f1_score_value = f1_score(true,pred,pos_label=1,zero_division=0,average="weighted")

    print("Precison value for training data: ",precision_value*100,"%")
    print("Recall value for training data: ",recall_value*100,"%")
    print("f1 score value for training data: ",f1_score_value*100,"%")

```

```python
#using training data
precision_recall_f1score(x_train_predict,y_train)
```
